```python
import os
import matplotlib.pyplot as plt
```

```python
path_train = "./train/"
path_valid = "./valid/"
```

```python
class_names = ["Albacore\ntuna",
               "Bigeye\ntuna",
               "Dolphinfish",
               "Opah",
               "No fish",
               "Other",
               "Shark",
               "Yellowfin\ntuna"]
```

```python
def count_class(class_name):
    return len([name for name in os.listdir(path_train+class_name)]) +\
            len([name for name in os.listdir(path_valid+class_name)])
```

```python
alb = count_class("ALB/")
det = count_class("BET/")
dol = count_class("DOL/")
lag = count_class("LAG/")
nof = count_class("NoF/")
other = count_class("OTHER/")
shark = count_class("SHARK/")
yft = count_class("YFT/")

class_counts = [alb, det, dol, lag, nof, other, shark, yft]
```

```python
plt.style.use("seaborn")
```

```python
plt.bar(class_names, class_counts)
plt.xlabel("Classes", fontsize=16)
plt.ylabel("Counts", fontsize=16)
plt.xticks(fontsize=12, rotation=45)
plt.savefig("./visualizations/fisheries_barplot.png", bbox_inches="tight", dpi=200)
#plt.xticks(np.arange(1,23,3), class_names, fontsize=12)
```
